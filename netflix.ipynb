```python
import numpy as np
import pandas as pd
import datetime as dt
```

```python
filename = f'/Users/Admin/Desktop/Data analytic/practice/netflix/netflix.csv'
data = pd.read_csv(filename)

data.head(5)
```

```python
# check for duplicate

data.duplicated().sum()
```

```python
# check for null value

for i in data.columns:
    null_rate = data[i].isna().sum() / len(data) * 100 
    if null_rate > 0 :
        print(f"{i} null rate: {round(null_rate,2)}%")

data.isnull().sum()
```

```python
# replace the missing value of country with most frequent country

data['country'] = data['country'].fillna(data['country'].mode()[0])
```

```python
# replace null values

data['cast'] = data['cast'].fillna('No Data')
data['director'] = data['cast'].fillna('No Data')

# drop other missing value

data.dropna(inplace=True)
```

```python
# change date format

data['date_added'] = pd.to_datetime(data['date_added'].str.strip())
```

```python
# add column month and year added

data['month_added']=data['date_added'].dt.month_name()
data['year_added'] = data['date_added'].dt.year
```
